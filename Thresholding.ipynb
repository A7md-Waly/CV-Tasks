```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
images = ["im1.jpg", "imm2.jpg", "im3.jpg", "im4.jpg", "im5.jpg", "im6.jpg", "im7.jpg", "im8.jpg"]
```

# Binary Thresholding

```python
for img_name in images:
    img = cv2.imread(img_name, cv2.IMREAD_GRAYSCALE)

    block_size = 11
    constant = 2
    adaptive_img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, constant)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(1, 2, 2), plt.imshow(adaptive_img, cmap='gray')
    plt.title('Adaptive Mean Thresholding'), plt.xticks([]), plt.yticks([])

    plt.show()
```

# ِِAdaptiveThreshold

```python
for img_name in images:
    img = cv2.imread(img_name, cv2.IMREAD_GRAYSCALE)

    block_size = 57
    constant = 1

    adaptive_img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, constant)

    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original Image'), plt.xticks([]), plt.yticks([])
    plt.subplot(1, 2, 2), plt.imshow(adaptive_img, cmap='gray'), plt.title('Adaptive Thresholding'), plt.xticks([]), plt.yticks([])
    plt.show()
```

# Gaussian Thresholding

```python
for img_name in images:
    img = cv2.imread(img_name, cv2.IMREAD_GRAYSCALE)
    gaussian_img = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, -2)
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(1, 2, 2), plt.imshow(gaussian_img, cmap='gray')
    plt.title('Gaussian Thresholding'), plt.xticks([]), plt.yticks([])

    plt.show()
```

# Otsu's Method

```python
for img_name in images:
    img = cv2.imread(img_name, cv2.IMREAD_GRAYSCALE)
    _, otsu_img = cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

    plt.figure(figsize=(10, 5))

    plt.subplot(1, 2, 1), plt.imshow(img, cmap='gray')
    plt.title('Original Image'), plt.xticks([]), plt.yticks([])

    plt.subplot(1, 2, 2), plt.imshow(otsu_img, cmap='gray')
    plt.title("Otsu's Method"), plt.xticks([]), plt.yticks([])

    plt.show()
```
