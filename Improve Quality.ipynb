```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = "ti.jpg"
image = cv2.imread(image_path)

blurred_image = cv2.GaussianBlur(image, (9, 9), 5)

gray_image = cv2.cvtColor(blurred_image, cv2.COLOR_BGR2GRAY)

increased = cv2.add(gray_image, (50, 90, 100 ,50))
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(increased, cv2.COLOR_BGR2RGB))
plt.title("Improved")


plt.show()
```
